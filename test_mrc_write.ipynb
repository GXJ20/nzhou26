```python
import mrcfile
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter
import numpy as np
import os
import pathlib
project_name = '20210803_monomer_Sustech'

i = 3435
j = 17
altered_dir = pathlib.Path(f'particleSeg/{project_name}/mrcs/')
original_dir = pathlib.Path(f'{project_name}/Extract/job481/')

altered_paths = sorted(altered_dir.glob("*.mrcs"))
original_paths = sorted(original_dir.glob('*/*.mrcs'))

image_name = altered_paths[i]
with mrcfile.open(original_paths[i], mode='r') as mrc:
    raw = mrc.data[j]
    raw = raw.astype(np.float32)
with mrcfile.open(altered_paths[i], mode='r') as mrc:
    masked = mrc.data[j]
    masked = masked.astype(np.float32)

print(image_name)
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.imshow(raw, cmap='gray')
plt.title('raw')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(gaussian_filter(raw, sigma=5), cmap='gray')
plt.title('low-pass filtered')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(masked, cmap='gray')
plt.title('subtracted')
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(gaussian_filter(masked, sigma=5), cmap='gray')
plt.title('subtracted low-pass filtered')
plt.axis('off')

```
